```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# IN COUNTRY-CODE DATAfRAME THERE ARE TWO COLUMNS "COUNTRY CODE" AND "COUNTRY" 
# AND IN ZOMATO DATAfRAME THERE IS ALSO WE CAN SEE THAT THERE IS "COUNTRY CODE" 
# SO,WE HAVE TO FIND THE RELATION BETWEEN THEM
# SO WE HAVE TO COMBINE BOTH DATAfRAMES
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left' )
```

```python
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# HERE WE GOT TO KNOWN THAT ZOMATO IS USED MOSTLY IN COUNTRY - "INDIA"
# SOO WE HAVE TO PLOT A PIE CHART OF TOP 3 COUNTRIES
```

```python
plt.pie(country_val[:3],labels=country_names[:3])
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct = '%1.2f%%')
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

# FROM ABOVE TABLE WE OBSERVE THE RATING FROM SCALE 0.0 TO 4.9

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6) #SIZE#
sns.barplot(x='Aggregate rating', y = 'Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y = 'Rating count',hue = 'Rating color',data=ratings)
```

# IN ABOVE GRAPH THE WHITE IS MARKED AS BLUE SO WE HAVE TO CHANGE IT

```python
sns.barplot(x='Aggregate rating', y = 'Rating count',hue = 'Rating color',palette=['white','red','orange','yellow','green','green'],data=ratings)
```
